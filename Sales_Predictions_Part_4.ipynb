# Part 4

```python
Sales['Item_Fat_Content'].value_counts().plot(kind='bar', title='Low Fat and Regular Item Counts')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=Sales, x='Item_Type', y='Item_Weight')
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Item Weight')
plt.title('Grocery Store Item Weights')
plt.show();
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
fig.set_figheight(4)
plt.plot(Sales['Item_Visibility'], Sales['Item_Outlet_Sales'], 'o' ,c='magenta')
axes.set_xlabel('Visibility', fontsize = 15)
axes.set_ylabel('Sales', fontsize = 15)
axes.set_title('Visibility Impact', fontsize = 16)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=Sales, x='Item_Type', y='Item_Outlet_Sales')
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Item Sales')
plt.title('Grocery Store Sales by Item')
plt.show();
```

```python
Mean_Sales= Sales.groupby('Outlet_Type')['Item_Outlet_Sales'].mean().reset_index(name = 'Mean Sales')
Mean_Sales
```

```python
plt.bar(Mean_Sales['Outlet_Type'], Mean_Sales['Mean Sales'])
sns.set()
plt.rcParams["figure.figsize"] = (10, 4)
plt.rcParams["xtick.labelsize"] = 13
plt.xlabel('Types of Outlet')
plt.ylabel('Mean Sales')
plt.title('Comparison of Average Sales for Outlet type')
```
